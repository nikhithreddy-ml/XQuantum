```python
high=[]
low=[]
for i in range(250):
    high.append(df.High[i])
    low.append(df.Low[i])
```

```python
high12=[]
for i in range(len(high)-60):
    new=[]
    for k in range(0+i,60+i):
        new.append(high[k])
        
    new.sort(reverse=True)
    high12.append(new[0])
a=high12[-1]   
for i in range(len(high12),250):
    high12.append(a)
```

```python
high24=[]
for i in range(len(high)-120):
    new=[]
    for k in range(0+i,120+i):
        new.append(high[k])
    new.sort(reverse=True)
    high24.append(new[0])
a=high24[-1]
for i in range(len(high24),250):
    high24.append(a)
```

```python
low12=[]
for i in range(len(low)-60):
    new=[]
    for k in range(0+i,60+i):
        new.append(low[k])
        
    new.sort()
    low12.append(new[0])
a=low[-1]   
for i in range(len(low12),250):
    low12.append(a)
```

```python
low24=[]
for i in range(len(low)-120):
    new=[]
    for k in range(0+i,120+i):
        new.append(low[k])
    new.sort()
    low24.append(new[0])
a=low24[-1]
for i in range(len(low24),250):
    low24.append(a)
```

```python
df["24 week trailing high"]=high24
```

```python
df["24 week trailing low"]=low24
```

```python
df["12 week trailing high"]=high12
```

```python
df["12 week trailing low"]=low12
```
